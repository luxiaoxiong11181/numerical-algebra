[1]. The difinition of operator norm is:
$$ 
\| \boldsymbol A \| _{\boldsymbol {\hat m \hat n}} \equiv \mathrm {max}_{x \neq 0, x \in \mathbb R ^n} \frac {\| \boldsymbol A \boldsymbol x \| _{\hat m}} {\| \boldsymbol x \| _{\hat n}} 
$$
This indicate that for all $ \boldsymbol x \in \mathbb R^n $ and operator norm $ \| \cdot \| _{\boldsymbol {\hat m \hat n}} $:
$$ 
\| \boldsymbol A \boldsymbol x \| _{\hat m} \le \| \boldsymbol A \| _{\boldsymbol {\hat m \hat n}} \| \boldsymbol x \| _{\hat n}
$$

if $ \| \cdot \| $ is vector two-norm and matrix Frobenius norm.For all $ \boldsymbol x \in \mathbb R^n $:
$$ 
\| \boldsymbol A \boldsymbol x \| _2 
= \sum _{i=0}^m \left( \sum _{j=0}^n a_{ij} x_j \right) ^2 \le \sum _{i=0}^m \left( \sum _{j=0}^n a_{ij}^2 \sum _{j=0}^n x_j^2 \right) 
= \| \boldsymbol A \| _F \| \boldsymbol x \| _2
$$

[2]. Use the result of [1], For all $ \boldsymbol x $:
$$ \| \boldsymbol A \boldsymbol B \boldsymbol x \| 
\le \| \boldsymbol A \| \| \boldsymbol B \boldsymbol x \| 
\le \| \boldsymbol A \| \| \boldsymbol B \boldsymbol x \| 
\le \| \boldsymbol A \| \| \boldsymbol B \| \| \boldsymbol x \| $$
Thus:
$$ 
\frac { \| \boldsymbol A \boldsymbol B \boldsymbol x \| } {\| \boldsymbol x \|} 
\le
\| \boldsymbol A \| \| \boldsymbol B \|
$$
it means:
$$ 
\| \boldsymbol A \boldsymbol B \| 
\le
\| \boldsymbol A \| \| \boldsymbol B \|
$$

    1 & 0 \\

$$
\| \boldsymbol A \|_{\mathrm F} 
= \sqrt 2
\neq 1
=
\frac {\| \boldsymbol x \|_{\mathrm F} }{\| \boldsymbol x \|_{\mathrm F} }
=
\max_{\boldsymbol y \in \mathbb R^n} \frac {\| \boldsymbol A \boldsymbol y \|_{\mathrm F} } {\| \boldsymbol y \|_{\mathrm F} }
$$

[4]. For $ \| \cdot \| _{\mathrm F} $, set $ \boldsymbol Q $ is a m-dim orthogonal or unitary and $ \boldsymbol Z $ is a n-dim orthogonal or unitary, for all $ \boldsymbol x \in \mathbb R^ m, \boldsymbol y \in \mathbb R^ n$:
$$
\| \boldsymbol Q \boldsymbol x \| _{\mathrm F} 
= \| \boldsymbol Q \boldsymbol x \| _2 
= \| \boldsymbol x \| _2  \\
\| \boldsymbol Z \boldsymbol y \| _{\mathrm F} 
= \| \boldsymbol Z \boldsymbol y \| _2 
= \| \boldsymbol y \| _2  \\
$$
Thus:
$$
\begin{alignat}{2}
\max_{\boldsymbol y \in \mathbb R^m} \frac {\| \boldsymbol Q \boldsymbol A \boldsymbol Z \boldsymbol y \| _F} {\| \boldsymbol y \|_F}
&=
\max_{\boldsymbol y \in \mathbb R^m} \frac {\| \boldsymbol Q \boldsymbol A \boldsymbol Z \boldsymbol y \| _F} {\| \boldsymbol Z^{-1} \boldsymbol Z  \boldsymbol y \|_F}
=
\max_{\boldsymbol z \in \mathbb R^m} \frac {\| \boldsymbol Q \boldsymbol A \boldsymbol z \| _F} {\| \boldsymbol z \|_F} \\
&=
\max_{\boldsymbol z \in \mathbb R^m} \frac {\| \boldsymbol Q \boldsymbol A \boldsymbol z \| _F} {\| \boldsymbol A \boldsymbol z \|_F} \frac {\| \boldsymbol A \boldsymbol z \|_F} {\| \boldsymbol z \|_F} 
=
\max_{\boldsymbol z \in \mathbb R^m} \frac {\| \boldsymbol A \boldsymbol z \|_F} {\| \boldsymbol z \|_F} \\
\end{alignat}
$$
$$
\begin{alignat}{2}
\max_{\boldsymbol y \in \mathbb R^m} \frac {\| \boldsymbol Q \boldsymbol A \boldsymbol Z \boldsymbol y \| _2} {\| \boldsymbol y \|_2}
&=
\max_{\boldsymbol y \in \mathbb R^m} \frac {\| \boldsymbol Q \boldsymbol A \boldsymbol Z \boldsymbol y \| _2} {\| \boldsymbol Z^{-1} \boldsymbol Z  \boldsymbol y \|_2}
=
\max_{\boldsymbol z \in \mathbb R^m} \frac {\| \boldsymbol Q \boldsymbol A \boldsymbol z \| _2} {\| \boldsymbol z \|_2} \\
&=
\max_{\boldsymbol z \in \mathbb R^m} \frac {\| \boldsymbol Q \boldsymbol A \boldsymbol z \| _2} {\| \boldsymbol A \boldsymbol z \|_2} \frac {\| \boldsymbol A \boldsymbol z \|_2} {\| \boldsymbol z \|_2} 
=
\max_{\boldsymbol z \in \mathbb R^m} \frac {\| \boldsymbol A \boldsymbol z \|_2} {\| \boldsymbol z \|_2} \\
\end{alignat}
$$

[5]. For all $ \boldsymbol x \in \mathbb R^ n$:
$$
\begin{alignat}{2}
\mathrm {max}_{\boldsymbol x \neq \boldsymbol 0} 
\frac {{\| \boldsymbol A \boldsymbol x \|}_{\infty}} {{\| \boldsymbol x \|}_{\infty}}
&=
\mathrm {max}_{\| \boldsymbol x \| _{\infty} = 1} 
{\| \boldsymbol A \boldsymbol x \|}_{\infty}
=
\mathrm {max}_{\mathrm {max}_{i \le n}  x_i = 1} 
{\| \boldsymbol A \boldsymbol x \|}_{\infty} \\
&=
\mathrm {max}_i  {\left| \left| \boldsymbol A 
\begin{bmatrix} \mathrm {sign} (a_{i0}) \\ \vdots \\ \mathrm {sign} (a_{in-1}) \\ \end{bmatrix}  \right| \right|}_{\infty}
= 
\mathrm {max}_i  \sum_{j=0}^{n-1} \left| a_{ij} \right| \\
&=
\| \boldsymbol A \| _{\infty}
\end{alignat}
$$

[6]. For all $ \boldsymbol x \in \mathbb R^ n$:
$$
\begin{alignat}{2}
\mathrm {max}_{\boldsymbol x \neq \boldsymbol 0} 
\frac {{\| \boldsymbol A \boldsymbol x \|}_1} {{\| \boldsymbol x \|}_1}
&=
\mathrm {max}_{\sum_j \| \boldsymbol x \|_1  = 1} 
{\| \boldsymbol A \boldsymbol x \|}_1
=
\mathrm {max}_{\sum_j \left| x_j \right|  = 1} 
{\| \boldsymbol A \boldsymbol x \|}_1 \\
&=
\mathrm {max}_{\sum_j \left| x_j \right|  = 1} 
\sum_i \left| \sum_j a_{ij} x_j \right|
= 
\mathrm {max}_{\mathrm {some} k, \left| x_k \right| = \delta (k, j)} 
\sum_i \left| \sum_j a_{ij} \right| \left| x_j \right| \\
&=
\| \boldsymbol A \| _{\infty}
\end{alignat}
$$

[7]. For all $ \boldsymbol x \in \mathbb R^ n$:
$$
\begin{alignat}{2}
\max_{\boldsymbol x \neq \boldsymbol 0} 
\frac {{\| \boldsymbol A \boldsymbol x \|} _2} {{\| \boldsymbol x \|}_2}
&=
\max_{\boldsymbol x \neq \boldsymbol 0} 
\frac {\sqrt {\boldsymbol x^* \boldsymbol A^* \boldsymbol A \boldsymbol x }} {\sqrt {\boldsymbol x^* \boldsymbol x}}
=
\max_{\boldsymbol x \neq \boldsymbol 0} 
\sqrt { \frac {\boldsymbol x^* \boldsymbol A^* \boldsymbol A \boldsymbol x } {\boldsymbol x^* \boldsymbol x} } \\
&=
\max_{\boldsymbol x \neq \boldsymbol 0} 
\sqrt { \frac {\boldsymbol x^* \boldsymbol Q^* \boldsymbol \Lambda \boldsymbol Q \boldsymbol x } {\boldsymbol x^* \boldsymbol Q^*  \boldsymbol Q \boldsymbol x} }
=
\max_{\boldsymbol y \neq \boldsymbol 0} 
\sqrt { \frac {\boldsymbol y^* \boldsymbol \Lambda \boldsymbol y } {\boldsymbol y^* \boldsymbol y} } \\
&=
\sqrt {\lambda_{\mathrm {max}}}
\end{alignat}
$$

[8]. For all $ \boldsymbol A $, set:
$$
\boldsymbol {A^* A} = \boldsymbol {Q^* \Lambda Q}
$$
Thus:
$$
\begin{alignat}{2}
\boldsymbol {A^{*T} A^T}
&=
\boldsymbol {A^{*T} Q^T Q^{*T} A^T}
=
\boldsymbol {(QA^*)^T (AQ^*)^T} \\
&=
\boldsymbol {(\Lambda (AQ^*) ^{-1})^T (AQ^*)^T}
=
\boldsymbol {(AQ^*) ^{T-1} \Lambda ^T (AQ^*)^T} \\
&=
\boldsymbol {(AQ^*) ^{T-1} \Lambda ^T (AQ^*)^T}
\end{alignat}
$$
Thus $ \boldsymbol {A^{*T} A^T} $ have the same eigenvalue matrix as $ \boldsymbol {A^* A} $, thus:
$$
\| \boldsymbol A^T \|_2 = \| \boldsymbol A \|_2
$$

[9]. For all $ \boldsymbol A : \boldsymbol A \boldsymbol A^* = \boldsymbol A^* \boldsymbol A $, exists unitary matrix $ \boldsymbol U $ and diagonal matrix $ \boldsymbol \Lambda $:
$$
\boldsymbol A = \boldsymbol U^* \boldsymbol \Lambda \boldsymbol U
$$
Thus:
$$
\begin{alignat}{2}
\max_{\boldsymbol x \neq \boldsymbol 0} 
\frac {{\| \boldsymbol A \boldsymbol x \|} _2} {{\| \boldsymbol x \|}_2}
&=
\max_{\boldsymbol x \neq \boldsymbol 0} \sqrt
{\frac { \boldsymbol {x^* A^* A x} } {\boldsymbol {x^* x} } }
=
\max_{\boldsymbol x \neq \boldsymbol 0} \sqrt
{\frac { \boldsymbol {x^* U^* \Lambda^* U U^* \Lambda U x} } {\boldsymbol {x^* x} } } \\
&=
\max_{\boldsymbol x \neq \boldsymbol 0} \sqrt
{\frac { \boldsymbol {x^* U^* \Lambda^* \Lambda U x} } {\boldsymbol {x^* x} } }
=
\max_{\boldsymbol x \neq \boldsymbol 0} \sqrt
{\frac { \boldsymbol {x^* U^* \Lambda^* \Lambda U x} } {\boldsymbol {x^* U^* U x} } } \\
&=
\max_{\boldsymbol y \neq \boldsymbol 0} \sqrt
{\frac { \boldsymbol {y^* \Lambda^* \Lambda y} } {\boldsymbol {y^* y} } }
=
\max_i \left| \lambda_i (\boldsymbol A) \right|
\end{alignat}
$$

[10]. If $ \boldsymbol A $ is n-by-n, then:
$$
\| \boldsymbol A \| _2^2 
=
\max_{\| \boldsymbol x \|_2 = 1} \| \boldsymbol {A x} \| _2
\le
\max_{\| \boldsymbol x \|_2 = 1} n^{1/2} \| \boldsymbol {A x} \| _1 
\le
n^{1/2} \| \boldsymbol A \| _1 \max_{\| \boldsymbol x \|_2 = 1} \| \boldsymbol x \| _1
\le
n^{1/2} \| \boldsymbol A \| _1
$$
$$
\| \boldsymbol A \| _1^2 
=
\max_{\| \boldsymbol x \|_1 = 1} \| \boldsymbol {A x} \| _1
\le
\max_{\| \boldsymbol x \|_1 = 1} n^{1/2} \| \boldsymbol {A x} \| _2
\le
n^{1/2} \| \boldsymbol A \| _2 \max_{\| \boldsymbol x \|_1 = 1} \| \boldsymbol x \| _2
\le
n^{1/2} \| \boldsymbol A \| _2
$$
Thus:
$$
n^{-1/2} \| \boldsymbol A \| _2 \le \| \boldsymbol A \|_1 \le n^{1/2} \| \boldsymbol A \| _2
$$

[11]. If $ \boldsymbol A $ is n-by-n, using [10]:
$$
n^{-1/2} \| \boldsymbol A^* \| _2 
\le
\| \boldsymbol A^* \| _1 
\le
n^{1/2} \| \boldsymbol A^* \|_2
$$

$$
\| \boldsymbol A^* \| _2 
=
\sqrt {\lambda _{\max} (\boldsymbol {A A^*})}
=
\sqrt {\lambda _{\max} (\boldsymbol {\overline A^* \overline A})}
=
\| \overline {\boldsymbol A} \| _2 
$$
$$
\| \boldsymbol A^* \| _1
=
\| \overline {\boldsymbol A} \| _{\infty}
$$
Thus:
$$
n^{-1/2} \| \overline {\boldsymbol A} \| _2 
\le
\| \overline {\boldsymbol A} \| _1 
\le
n^{1/2} \| \overline {\boldsymbol A} \|_2
$$
This equivalent to:
$$
n^{-1/2} \| \boldsymbol A \| _2 
\le
\| \boldsymbol A \| _1 
\le
n^{1/2} \| \boldsymbol A \|_2
$$

[13]. If $ \boldsymbol A $ is n-by-n:
$$
\| \boldsymbol A \| _2^2 
\le
\mathrm {tr} (\boldsymbol {A^* A})
=
\| \boldsymbol A \| _F^2

\mathrm {tr} (\boldsymbol {A^* A})

$$
\| \boldsymbol A \| _1 
\le
\| \boldsymbol A \| _F
