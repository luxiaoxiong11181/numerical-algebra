性质 1

证：

 因为$ \lVert A\rVert = \max\limits_{x\neq0}\frac{\lVert Ax\rVert}{\lVert x\rVert}$
 
 
 所以 当 $x\neq0$ 时，有$\frac{\lVert Ax\rVert}{\lVert x\rVert}\leq\lVert A\rVert$
 
 
 所以$\lVert Ax\rVert\leq\lVert A\rVert\lVert x\rVert$
 
 
 当 $x = 0$ 时，$\lVert Ax\rVert=\lVert A\rVert\lVert x\rVert$
 
 综上，$\lVert Ax\rVert\leq\lVert A\rVert\lVert x\rVert$
 
 所证成立

性质 4

 证明    首先证明$\forall A \in C^{m\times n}$成立
$$\left \| A \right \|_F^2 = tr(A^HA) = tr(AA^H)
$$

 事实上 $A^HA$的$i$行$i$列的元素为
$$\sum_{k=1}^m \overline{a_{ki}}a_{ki} = \sum_{k=1}^m\left|a_{ki}\right|^2
$$

 因此
$$tr(A^HA) = \sum_{i=1}^n\sum_{k=1}^m\left|a_{ki}\right|^2 = \left \| A \right \|_F^2
$$

于是
$$\left \| A \right \|_F^2 = \left \| A^H \right \|_F^2 = tr((A^H)^HA^H) = tr(AA^H)
$$

下证   当Q和Z是酉阵时（正交阵同理可证），对Frobenius范数满足
$$\left \| QAZ \right \|_F=\left \| A \right \|_F
$$

因为
$$\left \| QAZ \right \|_F^2 = tr((QAZ)^H(QAZ)) = tr(Z^HA^HQ^HQAZ) = tr(Z^HA^HAZ) = tr(A^HA) = \left \| A \right \|_F^2
$$

故
$$\left \| QAZ \right \|_F=\left \| A \right \|_F
$$

证明   当Q和Z是酉阵时（正交阵同理可证），对由$\left \| \cdot  \right \|_2$导出的算子范数成立
$$\left \| QAZ \right \|_2=\left \| A \right \|_2
$$

 因为
$$\left \| QAZ \right \|_2 = \sqrt{\lambda_{max}((QAZ)^H(QAZ))} =\sqrt{\lambda_{max}(Z^HA^HQ^HQAZ)} = \sqrt{\lambda_{max}(Z^HA^HAZ)} = \sqrt{\lambda_{max}(A^HA)} = \left \| A \right \|_2
$$

 故
$$\left \| QAZ \right \|_2=\left \| A \right \|_2
$$

性质 8

 证明 首先证明$\qquad$若 A 和 B都是$n\times n$阶矩阵，则AB和BA有相同的特征值
 $\quad$设$\lambda$是AB的一个特征值，$X$是对应于$\lambda$的一个特征向量.则$$\quad(AB)X = \lambda X\quad \quad$$等式两边左乘B.$\quad$得到
$$BA(BX) = \lambda(BX)
$$
 当$BX = 0$时，$\lambda X = A(BX) = 0$,而$X\neq 0$,则$\lambda = 0$,于是
 $$\left| AB -0E \right| = \left| AB \right| = \left| BA \right| = \left| BA -0E \right| = 0
$$
 所以$\lambda = 0$ 是BA的特征值
 当$BX \neq 0$时，由定义有非零向量$BX$就是$BA$对应于特征值$\lambda$的特征向量，所以$\lambda$也是$BA$的特征值

性质 9  $$
{\lVert A\rVert}_2 = \max\limits_i|\lambda_i(A)|,当A是正规阵，即AA^* = A^*A时成立
$$
证明： $A$ 是正规阵，必然存在酉阵 $Q$ 满足 $Q^\prime *A*Q = D$ , $D$ 为对角阵且对角线的元素都是 $A$ 的特征值。记 $D = diag{{\lambda_1, \lambda_2, \cdots, \lambda_n}}$ 满足 $|\lambda_1|\geq|\lambda_2|\geq\cdots\geq|\lambda_n|$ ，则 $|\lambda_1|$ 为 $A$ 的最大的特征值。
$$
\quad
$$ 
令 $x_1$为 $\lambda_1$ 对应的右特征向量，满足 $A*x_1 = \lambda_1*x_1$ ,必然有： $\frac{{\lVert A*x_1\rVert}_2}{{\lVert x_1\rVert}_2} = |\lambda_1|{\geq\lVert A\rVert}_2$ 
$$
\quad
$$
令 $y_1$ 为 $\lVert A\rVert_2$ 对应的单位向量，即： $\lVert y_1\rVert_2 = 1$ 且 $\lVert A\rVert_1 = \lVert A*y_1\rVert_2$ . $y_1$可以被 $Q$ 线性表出为 $y_1 = Q*z_1$ ,且 $z_1$ 也为单位向量。那么不难得出：
$$
\lVert A\rVert_2 = \lVert A*y_1\rVert_2 = \lVert A*Q*z_1\rVert_2 = \lVert Q*z_1\rVert_2 \leq|\lambda_1|
$$
即： $\lVert A\rVert_2 = \max\limits_i|\lambda_i(A)|$

性质 10

证：
由$\lVert x \rVert _2 \le\lVert  x \rVert _1\le n^{1/2}\lVert x \rVert_2$ 得

$\lVert Ax \rVert _1\le n^{1/2}\lVert Ax \rVert_2\le n^{1/2}\lVert A \rVert_2\lVert x \rVert_2\le n^{1/2}\lVert A \rVert_2\lVert x \rVert_1$

所以$\lVert  A \rVert _1\le n^{1/2}\lVert A \rVert_2$

$\lVert Ax \rVert _2\le \lVert Ax \rVert_1\le \lVert A \rVert_1\lVert x \rVert_1\le n^{1/2}\lVert A \rVert_1\lVert x \rVert_2$

所以$\lVert  A \rVert _2\le n^{1/2}\lVert A \rVert_1$

所以$n^{-1/2}\lVert A \rVert _2 \le\lVert  A \rVert _1\le n^{1/2}\lVert A \rVert_2$

$\lVert Ax \rVert _\infty \le\lVert  Ax \rVert _2\le\lVert  A \rVert _2\lVert  x \rVert _2\le n^{1/2}\lVert  A \rVert _2\lVert x \rVert_\infty$

所以$\lVert A\rVert_\infty \leq n^{1/2}\lVert A\rVert_2$

$\lVert Ax \rVert _2 \le n^{1/2}\lVert  Ax \rVert _\infty\le n^{1/2}\lVert  A \rVert _\infty\lVert x \rVert_\infty\le n^{1/2}\lVert  A \rVert _\infty\lVert x \rVert_2$

所以$n^{-1/2}\lVert A\rVert_2 \leq\lVert A\rVert_\infty $

所以 $n^{-1/2}\lVert A\rVert_2 \leq\lVert A\rVert_\infty \leq n^{1/2}\lVert A\rVert_2$

性质 12

证：
由$\lVert x \rVert _\infty \le\lVert  x \rVert _1\le n\lVert x \rVert_\infty$ 

$\lVert  Ax \rVert _1\le n\lVert Ax \rVert_\infty\le n\lVert A \rVert_\infty\lVert x \rVert_\infty\le n\lVert A \rVert_\infty\lVert x \rVert_1$

所以$\lVert  A \rVert _1\le n\lVert A \rVert_\infty$

$\lVert Ax \rVert_\infty\le \lVert Ax \rVert_1\le \lVert A \rVert_1\lVert x \rVert_1\le n\lVert A \rVert_1\lVert x \rVert_\infty$

所以$n^{-1}\lVert A\rVert_\infty\le\lVert A\rVert_1$

所以$n^{-1}\lVert A\rVert_\infty\le\lVert A\rVert_1\le n\lVert A \rVert_\infty$

\lVert A\rVert_1^2 = \max\limits_i\sum_{j = 1}^n |a_{ij}|

因此
$$
\lVert A\rVert_1^2\leq\sum_{i=1}^n|a_{i1}|^2 +\sum_{i=1}^n|a_{i2}|^2 + \cdots +\sum_{i=1}^n|a_{in}|^2 = \sum_{i = 1}^n\sum_{j = 1}^n|a_{ij}|^2 = \lVert A\rVert_F^2

$$
\quad

$$

$$
即： $\lVert A\rVert_F \leq n^{\frac{1}{2}}\lVert A\rVert_2$
$$
\quad
$$
所以
$$
\lVert A\rVert_1 \leq\lVert A\rVert_F \leq n^{\frac{1}{2}}\lVert A\rVert_2
