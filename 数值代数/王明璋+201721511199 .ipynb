1.$  证明:   \left\|Ax\right\|\le\|A\|\cdot\|x\|,对一个向量范数及其对应的算子范数，或向量2范数和矩阵Frobenius范数成立$

  $ 证明：     $
 $$\left\|A\right\|\ =\max_{x\ne 0}\frac{\|Ax\|}{\|x\|}$$ 
  $ 当{ x\ne 0}时，    \frac{\|Ax\|}{\|x\|}\le\left\|A\right\| ,从而 \left\|Ax\right\|\le\|A\|\cdot\|x\|$
  
  $ 当x=0时，\left\|Ax\right\|\le\|A\|\cdot\|x\|显然成立$
  
2.$  证明:   \left\|AB\right\|\le\|A\|\cdot\|B\|,对一个任何算子范数或Frobenius范数成立，换言之，任何算子范数和它本身相互相容的    $

  $ 证明：     $
$$\left\|AB\right\|\ =\max_{x\ne 0}\frac{\|(AB)x\|}{\|x\|}=\max_{x\ne 0}\frac{\|A(Bx)\|}{\|x\|}\le\max_{x\ne 0}\frac{\left\|A\right\|\|Bx\|}{\|x\|}=\|A\|\cdot\max_{x\ne 0}\frac{\|Bx\|}{\|x\|}=\|A\|\|B\|   $$

3.$  证明:   最大范数和Frobenius范数不是算子范数 $
 
 $  证明：    $
 $假设Frobenius范数是算子范数 $
 $$则\left\|A\right\|_{F}=\max_{\|x\|=1}\|Ax\| $$
 $$取A=I,则\left\|I\right\|_{F}=\max_{\|x\|=1}\|Ix\|=1\quad\  (1)$$
 实际上
 $$\left\|I\right\|_{F}=(\sum_{i,j}^n|a_{ij}|^2)^\frac{1}{2}=n\quad\  (2)$$
(1)与（2）矛盾
所以Frobenius范数不是算子范数

4.$  \|QAZ\|=\|A\|,当Q和Z是正交阵或酉阵时.对弗罗贝尼乌斯范数及由\|\cdot\|_{2}导出的算子范数成立.而这实际上正好是毕达哥拉斯定理.$
 $  证明：    $
 $$ (QAZ)^*(QAZ)=(Z^*A^*Q^*)(QAZ)=Z^*(A^*A)Z,$$
 $${\lambda}((QAZ)^*(QAZ))={\lambda}(A^*A),$$
 $ 所以 \|QAZ\|_{2}^2=max{\lambda}((QAZ)^*(QAZ))=max{\lambda}(A^*A)=\|A\|_{2}^2$
 $$\|A\|_{F}^2=\sum_{i,j}^n a_{ij}^2=tr(A^*A)=\sum_{i=1}^n(a_{i}^2),其中a_{1}，a_{2}，\cdots,a_{n}是A^*A的特征值 $$
 $$\|A\|_{F}^2=\sum_{i=1}^n a_{i}^2=\sum_{i=1}^n{\lambda}_{i}^2((QAZ)^*(QAZ))=\|QAZ\|_{F}^2$$
 
 5.$$  \|A\|_{\infty}\equiv\max_{x\ne 0}\frac{\|Ax\|_{\infty}}{\|x\|_{\infty}}=\max_{i}\sum_{j}|a_{ij}|=最大的绝对值的行和   $$
 
 $ \quad  证明：  $   
 $$设 x=(x_{1},x_{2},\cdots,x_{n})^T\ne 0,设A\ne 0 , u=\max_{i}\sum_{j}|a_{ij}|$$ 
 $$记t=\|x\|_{\infty}=\max_{i}|x_{i}| $$
 $$则\|Ax\|_{\infty}=\max_{i}\sum_{j}|a_{ij}x_{j}|\le\max_{i}\sum_{j}|a_{ij}||x_{j}|\le t\max_{i}\sum_{j}|a_{ij}| $$
 $$所以\frac{\|Ax\|_{\infty}}{\|x\|_{\infty}}\le u=\max_{i}\sum_{j}|a_{ij}|$$
 $$u=\sum_{j=1}^n|a_{i_{0}j}|,取x_{0}=(x_{1},x_{2},\cdots,x_{n})^T,$$
 $$其中 若a_{i_{0}j}=0,则取x_{j}=0,若a_{i_{0}j}>0,则取x_{j}=1,a_{i_{0}j}<0,则取x_{j}=-1,$$
 $$(j=1,2,\cdots,n),\|x_{0}\|_{\infty}=1$$
 $$Ax_{0}的第i_{0}个分量为\sum_{j=1}^n|a_{i_{0}j}x_{j}|=\sum_{j=1}^n|a_{i_{0}j}|,
\|Ax_{0}\|_{\infty}=\max_{i}|\sum_{j=1}^na_{ij}x_{j}|=\sum_{j=1}^n|a_{i_{0}j}|=u ,$$
 $$所以\|A\|_{\infty}\equiv\max_{x\ne 0}\frac{\|Ax\|_{\infty}}{\|x\|_{\infty}}=\max_{i}\sum_{j}|a_{ij}|=最大的绝对值的行和 $$
 
 6.$$  \|A\|_{1}\equiv\max_{x\ne 0}\frac{\|Ax\|_{1}}{\|x\|_{1}}=\max_{j}\sum_{i}|a_{ij}|=最大的绝对值的列和   $$
 $ \quad  证明：  $ 
 $$设w=\max_{j}\sum_{i}|a_{ij}|,\quad \|x\|_{1}=1,A=[a_{1},a_{2},\cdots,a_{n}],则有w=\max_{j}\|a_{j}\|_{1},$$
 $$设 x=(x_{1},x_{2},\cdots,x_{n})^T,$$
 $$则\|Ax\|_{1}=\|x_{1}a_{1}+x_{2}a_{2}+\cdots+x_{n}a_{n}\|_{1}\le|x_{1}|\|a_{1}\|+|x_{2}|\|a_{2}\|+\cdots+|x_{n}|\|a_{n}\| \qquad\le(|x_{1}|+|x_{2}| +\cdots+|x_{n}|)w$$
 $$另一方面,设\sum_{i}|a_{ir}|=w,取 x_{r}=(0,0,\cdots,1,\cdots,0),(第r项为1，其余为0)，$$
$$  则\|x_{r}\|_{1}=1 且 \|Ax_{r}\|_{1}=\sum_{i}|a_{ir}|=w,$$
$$因此\|A\|_{1}\equiv\max_{x\ne 0}\frac{\|Ax\|_{1}}{\|x\|_{1}}=\max_{j}\sum_{i}|a_{ij}|$$

 7.$$ \|A\|_{2}\equiv\max_{x\ne 0}\frac{\|Ax\|_{2}}{\|x\|_{2}}=\lambda_{max}(A^*A),\lambda_{max}=最大的特征值 $$
 $ \quad  证明：  $ 
 $$因为A^*A是埃尔米特矩阵，所以存在一个特征分解A^*A=Q\Lambda Q^*,其中Q是酉阵,(列是特征向量) $$
 $$\Lambda=diag(\lambda_{1},\cdots,\lambda_{n})是包含特征值的对角阵，而特征值都是实的,所有的\lambda_{i}\ge 0,因为若某个\lambda 是负的，则取q为它的特征向量，有0\le \|Aq\|_{2}^2=q^TA^TAq=q^T\lambda q=\lambda\|q\|_{2}^2<0,矛盾 $$
 $$因此\|A\|_{2}\equiv\max_{x\ne 0}\frac{\|Ax\|_{2}}{\|x\|_{2}}= \max_{x\ne 0}\frac{(x^*A^*Ax)^\frac{1}{2}}{\|x\|_{2}}
 =\max_{x\ne 0}\frac{(x^*Q\Lambda Q^*x)^\frac{1}{2}}{\|x\|_{2}}=\max_{x\ne 0}\frac{((Q^*x)^*\Lambda Q^*x)^\frac{1}{2}}{\|Q^*x\|_{2}}=\max_{y\ne 0}\frac{(y^*\Lambda y)^\frac{1}{2}}{\|y\|_{2}}=\max_{y\ne 0}(\frac{\sum y_{i}^2\lambda_{i}}{y_{i}^2})^\frac{1}{2}\le max_{y\ne 0}\sqrt{\lambda_{max}}\sqrt{(\frac{\sum y_{i}^2}{\sum y_{i}^2})}=\sqrt{\lambda_{max}}$$
 
 8.$$ \|A\|_{2}=\|A^T\|_{2}  $$
 $ \quad  证明：  $ 
 $$设\lambda 是AA^T的特征值,有AA^Tx=\lambda x (x\ne 0),所以(A^TA)(A^Tx)=\lambda (A^Tx) (x\ne 0), $$
 $$(i)\lambda\ne 0,则A^Tx\ne 0,可知\lambda 也是A^TA的特征值 $$
 
 $$(ii)若 \lambda= 0,AA^T有零特征值,有|0E-A^TA|=|-A^TA|=|-A^T||A|=|A||A^T|=|0E-AA^T|=0,
 所以AA^T的特征值全是A^TA的特征值，同理
 A^TA的特征值是AA^T的特征值，AA^T与A^TA的特征值相同$$
 $$\|A\|_{2}=\sqrt{\lambda_{max}(A^*A)}=\sqrt{\lambda_{max}(AA^*)}=\|A^T\|_{2}  $$
 
 
 9.$$ \|A\|_{2}=\max_{i}\lambda_{i}|(A)|,当A是正规阵，即AA^*=A^*A时成立. $$
 
 $ \quad  证明：  $ 
 $$ \quad A是正规阵,必然存在酉阵Q满足Q^TAQ=D,D为对角阵且对角线的元素都是A的特征值，$$
 $$记D=diag(\lambda_{1},\cdots,\lambda_{n})满足|\lambda_{1}|\ge|\lambda_{2}|\ge\cdots\ge|\lambda_{n}|,$$
$$ 则|\lambda_{1}|为A的最大特征值,
 令x_{l}为\lambda_{1}对应的特征向量，满足Ax_{l}=\lambda_{1}x,必然有\frac{\|Ax_{1}\|_{2}}{\|x_{1}\|_{2}}|\lambda_{1}|\ge\|A\|_{2} $$
 $$令y_{l}为\|A\|_{2}对应的单位向量,即\|y_{l}\|_{2}=1且\|A\|_{2}=\|Ay_{l}\|_{2},y_{l}可以被Q线性表出为y_{l}=Qz_{l},\quad z_{l}为单位向量，$$
 $$那么不难得出\|A\|_{2}=\|Ay_{l}\|_{2}=\|AQz_{l}\|_{2}=\|Qz_{l}\|_{2}\le|\lambda_{1}|
 \|A\|_{2}=\max_{i}|\lambda_{i}(A)| $$ 
 
 10.$$ 若A是n*n矩阵,则n^\frac{-1}{2}\|A\|_{2}\le\|A\|_{1}\le n^\frac{1}{2}\|A\|_{2}. $$
  $ \quad  证明：  $ 
  $ 由向量范数的等价性,存在常数，使对任意x有
  c_{1}\|Ax\|_{2}\le\|Ax\|_{1}\le c_{2}\|Ax\|_{2},
  c_{2}\|x\|_{2}\le\|x\|_{1}\le c_{1}\|x\|_{2},$
  $ \quad  \frac{c_{1}}{c_{2}}\frac{\|Ax\|_{2}}{\|x\|_{2}}\le\frac{\|Ax\|_{2}}{\|x\|_{2}}\le\frac{c_{2}}{c_{1}}\frac{\|Ax\|_{2}}{\|x\|_{2}},
  \frac{c_{1}}{c_{2}}=n^\frac{-1}{2},\frac{c_{2}}{c_{1}}=n^\frac{1}{2},
  则有n^\frac{-1}{2}\frac{\|Ax\|_{2}}{\|x\|_{2}}\le\frac{\|Ax\|_{2}}{\|x\|_{2}}\le n^\frac{1}{2}\frac{\|Ax\|_{2}}{\|x\|_{2}}，$
 $$ n^\frac{-1}{2}\max_{x\ne 0}\frac{\|Ax\|_{2}}{\|x\|_{2}}\le\max_{x\ne 0}\frac{\|Ax\|_{2}}{\|x\|_{2}}\le n^\frac{1}{2}\max_{x\ne 0}\frac{\|Ax\|_{2}}{\|x\|_{2}}$$
 $$n^\frac{-1}{2}\|A\|_{2}\le\|A\|_{1}\le n^\frac{1}{2}\|A\|_{2}$$
  
 11.$$ 若A是n*n矩阵,则n^\frac{-1}{2}\|A\|_{2}\le\|A\|_{\infty}\le n^\frac{1}{2}\|A\|_{2}. $$
 $ \quad  证明：  $ 
 $$因为\|x\|_{\infty}\le\|x\|_{2}\le\sqrt n\|x\|_{\infty}$$
 $$所以\|Ax\|_{\infty}\le\|Ax\|_{2}\le\|A\|_{2}\|x\|_{2}\le\sqrt n\|A\|_{2}\|x\|_{\infty},$$
 $$再由无穷范数的定义得\|A\|_{\infty}\le n^\frac{1}{2}\|A\|_{2}$$
 $$另一方面,n^\frac{-1}{2}\|A\|_{2}\le\|Ax\|_{\infty}\|A\|_{\infty}\|x\|_{\infty}\le\|A\|_{\infty}\|x\|_{2}$$
$$再由2-范数的定义得n^\frac{-1}{2}\|A\|_{2}\le\|A\|_{\infty}$$

12.$$ 若A是n*n矩阵,则n^{-1}\|A\|_{\infty}\le\|A\|_{1}\le n\|A\|_{\infty}. $$
 $  \quad  证明：  $ 
 $$因为\|x\|_{\infty}\le\|x\|_{1}\le n\|x\|_{\infty}$$
 $$所以n^{-1}\|Ax\|_{\infty}\le n^{-1}\|Ax\|_{1}\le n^{-1}\|A\|_{1}\|x\|_{1}\le \|A\|_{1}\|x\|_{\infty},$$
 $$再由无穷范数的定义得n^{-1}\|A\|_{\infty}\le\|A\|_{1}$$
 
 $$\|Ax\|_{1}\le n\|Ax\|_{\infty}\le n\|A\|_{\infty}\|x\|_{\infty}\le n\|A\|_{\infty}\|x\|_{1}$$
$$再由1-范数的定义得\|A\|_{1}\le n\|A\|_{\infty}$$

13.$$ 若A是n*n矩阵,则\|A\|_{1}\le\|A\|_{F}\le n^{1/2}\|A\|_{2}. $$
   
 $ \quad  证明：  $
 $$ \|A\|_{2}^2=\lambda_{1}(A^*A)\ge n^{-1}[\lambda_{2}(A^*A)+\lambda_{max}(A^*A)+\cdots+\lambda_{n}(A^*A)]= n^{-1}tr(A^*A)=n^{-1}\|A\|_{F}^2$$
 $$ \|A\|_{F}\le n^{1/2}\|A\|_{2} $$
